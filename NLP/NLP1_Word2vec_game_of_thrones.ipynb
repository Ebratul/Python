```python
import os
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

folder_path = '/content/game-of-thrones-books'
value = []

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        f = open(os.path.join(folder_path, filename), 'r', encoding='utf-8', errors='ignore')
        corpus = f.read()
        f.close()  # ফাইল ম্যানুয়ালি বন্ধ করো
        for sent in sent_tokenize(corpus):
            value.append(simple_preprocess(sent))

print(f"Total sentences tokenized: {len(value)}")
print(value[:3])
```

```python
flat_list = []

for sublist in value:
    for item in sublist:
        flat_list.append(item)
# flat_list
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(flat_list, columns = ['word'])
df.head(5)
```

```python
df['word'].str.lower()
```

# remove punctuation

```python
import string, time
```

```python
def remove_punctuatioN(text):
  return text.translate(str.maketrans('', '', string.punctuation))
```

```python
df['word'] = df['word'].apply(remove_punctuatioN)
```

#lamatization

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lamatizer = WordNetLemmatizer()
```

```python
df['word'] = df['word'].apply(lambda x:lamatizer.lemmatize(x))
```

```python
df['word']
```

# remove stopword

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df['word'] = df[df['word'].apply(lambda x: x not in stop_words)].reset_index(drop=True)
df = df.dropna().reset_index(drop=True)
df
```

```python
# df to convert list
word = df['word'].apply(lambda x:[x]).tolist()
word
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,

)
```

```python
model.build_vocab(word)
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.train(value, total_examples=model.corpus_count, epochs=model.epochs)
```
