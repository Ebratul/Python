```python
#create own spacy


from spacy.tokens import Span

s1 = Span(doc1, 0, 1, label = "ORG")
s2 = Span(doc1, 5, 6, label= "ORG")

doc1.set_ents([s1, s2], default = "unmodified")
```

```python
for ent in doc1.ents:
  print(ent.text, "|" , ent.label_)
```

```python
#create own spacy

import spacy
from spacy.pipeline import EntityRuler

nlp = spacy.load("en_core_web_sm")
ruler = nlp.add_pipe("entity_ruler", before="ner")

patterns = [
    {"label": "COUNTRY", "pattern": "India"},
    {"label": "CITY", "pattern": "Delhi"},
    {"label": "STATE", "pattern": "Gujarat"},
    {"label": "STATE", "pattern": "Tamilnadu"},
    {"label": "STATE", "pattern": "Andhrapradesh"},
    {"label": "STATE", "pattern": "Assam"},
    {"label": "STATE", "pattern": "Bihar"}
]
ruler.add_patterns(patterns)

text = """Kiran want to know the famous foods in each state of India. So, he opened Google and search for this question. Google showed that
in Delhi it is Chaat, in Gujarat it is Dal Dhokli, in Tamilnadu it is Pongal, in Andhrapradesh it is Biryani, in Assam it is Papaya Khar,
in Bihar it is Litti Chowkha and so on for all other states"""

doc = nlp(text)

for ent in doc.ents:
    print(ent.text, '-->>', ent.label_)
```
