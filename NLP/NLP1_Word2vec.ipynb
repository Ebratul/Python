```python
od.download(r"https://www.kaggle.com/datasets/leadbest/googlenewsvectorsnegative300")
```

```python
model = KeyedVectors.load_word2vec_format(r'/content/googlenewsvectorsnegative300/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
model['man'].shape
```

```python
model['student']
```

```python
model.most_similar('man')
```

```python
model.most_similar('facebook')
```

```python
model.similarity('men','woman')
```

```python
model.similarity('bangladesh', 'dhaka')
```

```python
model.similarity('man', 'python')
```

```python
model.doesnt_match(['java', 'python', 'cricket'])
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```

```python
vec = model['INR'] - model['Bangladesh'] + model['UK']
model.most_similar([vec])
```
